```python
# Binary Triangle

n = int(input("Enter the number of rows: "))

for i in range(n):
    # Print leading spaces for centering
    for k in range(n - i - 1):
        print(' ', end='')

    # Print alternating 0s and 1s for the current row
    for j in range(i + 1):
        if (i + j) % 2 == 0:
            print('0', end=' ')
        else:
            print('1', end=' ')
    print()
```
