```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(19680801)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
yticks = [3,2,1,0]

for y in yticks:
    xs = np.arange(20)
    ys = np.random.rand(20)
    ax.bar(xs,ys,zs=y,zdir='y')

ax.set_yticks(yticks)
plt.figsize=(20,20)
plt.show()
```
